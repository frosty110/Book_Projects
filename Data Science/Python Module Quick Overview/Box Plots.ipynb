```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
sns.set_style("whitegrid")
tips = sns.load_dataset("tips")
tips.head()
```

```python
ax = sns.boxplot(x=tips["total_bill"])
```

```python
# verical box plot grouped by a categorical value.
ax = sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# nested grouping by 2 categorical variables
ax = sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
```

```python
# swarmplot to show the datapoints on top of the boxes
ax = sns.boxplot(x="day", y="total_bill", data=tips)
ax = sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
```

```python
# use factorplot, boxplot, and facet grid to show additional categorical variables.
g = sns.factorplot(x="sex", y="total_bill",
                    hue="smoker", col="time",
                    data=tips, kind="box",
                    size=4, aspect=.7);
```

```python
 ax = sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
# splot violins to compare with the hue variable. Scale enforces the number of observations
ax = sns.violinplot(x="day", y="total_bill", hue="smoker",
                     data=tips, palette="muted", scale="count", split=True, inner="stick")
```

```python
# Scale the density relative to the counts across all bins:
ax = sns.violinplot(x="day", y="total_bill", hue="smoker",
                     data=tips, palette="muted", scale="count", split=True, inner="stick", scale_hue=False,)
```

```python
# use factorplot() to combine with violinplot() and a facetGrid to allow additional categorical variable. factorplot() ensures synchronization of variables order across facets
g = sns.factorplot(x="sex", y="total_bill",
                    hue="smoker", col="time",
                    data=tips, kind="violin", split=True,scale="count",scale_hue=False,
                    size=4, aspect=.7);
```
